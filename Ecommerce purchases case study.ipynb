## 8. How many people have French 'fr' as their language

```python
print('Total french speaking people are',len(data[data['Language']=='fr']))
```

## 9. Job title contains engineer

```python
print("The number of people with job title as Engineer are",len(data[data['Job'].str.contains('engineer',case=False)]))
```

## 10. Find the email of a person with the 132.207.160.22 Ip Address

```python
print("The email address with the given IP Adress is ", data[data['IP Address']=='132.207.160.22']['Email'])
```

## 11. How many people have Mastercard as their credit card provider and made a purchase above 50?

```python
print("The number of people with given conditions are ",len(data[(data['CC Provider']=='Mastercard') & (data['Purchase Price']>50)]))
```

## 13. How many people purchase during the AM and How many people purchase during PM?

```python
print("The Number of people purchase during AM are" ,len(data[data['AM or PM']=='AM']))
print("The Number of people purchase during PM are" ,len(data[data['AM or PM']=='PM']))
                    ##OR
data['AM or PM'].value_counts()
```

## 14.How many people have a credit card that expires in 2020?

```python
len(data[data['CC Exp Date'].apply(lambda x:x[3:]=='20')])
```
